```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import MeCab
import re

%matplotlib inline

# å²©æ³¢å›½èªè¾æ›¸ã®ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰
import urllib.request
import zipfile
# URLã‚’æŒ‡å®š
url = "https://storage.googleapis.com/tutor-contents-dataset/6050_stock_price_prediction_data.zip"
save_name = url.split('/')[-1]
# ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã™ã‚‹
mem = urllib.request.urlopen(url).read()
# ãƒ•ã‚¡ã‚¤ãƒ«ã¸ä¿å­˜
with open(save_name, mode='wb') as f:
    f.write(mem)
# zipãƒ•ã‚¡ã‚¤ãƒ«ã‚’ã‚«ãƒ¬ãƒ³ãƒˆãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã«å±•é–‹ã™ã‚‹
zfile = zipfile.ZipFile(save_name)
zfile.extractall('.')


# MeCabã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã®ä½œæˆï¼å¼•æ•°ã‚’ç„¡æŒ‡å®šã«ã™ã‚‹ã¨IPAè¾æ›¸ã«ãªã‚Šã¾ã™ï¼
m = MeCab.Tagger('')

# ãƒ†ã‚­ã‚¹ãƒˆã‚’å½¢æ…‹ç´ è§£æã—è¾æ›¸ã®ãƒªã‚¹ãƒˆã‚’è¿”ã™é–¢æ•°
def get_diclist(text):
    parsed = m.parse(text)      # å½¢æ…‹ç´ è§£æçµæœï¼ˆæ”¹è¡Œã‚’å«ã‚€æ–‡å­—åˆ—ã¨ã—ã¦å¾—ã‚‰ã‚Œã‚‹ï¼‰
    lines = parsed.split('\n')  # è§£æçµæœã‚’1è¡Œï¼ˆ1èªï¼‰ã”ã¨ã«åˆ†ã‘ã¦ãƒªã‚¹ãƒˆã«ã™ã‚‹
    lines = lines[0:-2]         # å¾Œã‚2è¡Œã¯ä¸è¦ãªã®ã§å‰Šé™¤
    diclist = []
    for word in lines:
        l = re.split('\t|,',word)  # å„è¡Œã¯ã‚¿ãƒ–ã¨ã‚«ãƒ³ãƒã§åŒºåˆ‡ã‚‰ã‚Œã¦ã‚‹ã®ã§
        d = {'Surface':l[0], 'POS1':l[1], 'POS2':l[2], 'BaseForm':l[7]}
        diclist.append(d)
    return(diclist)

# å½¢æ…‹ç´ è§£æçµæœã®å˜èªã”ã¨ã®dictãƒ‡ãƒ¼ã‚¿ã«PNå€¤ã‚’è¿½åŠ ã™ã‚‹é–¢æ•°
def add_pnvalue(diclist_old, pn_dict):
    diclist_new = []
    for word in diclist_old:
        base = word['BaseForm']        # å€‹ã€…ã®è¾æ›¸ã‹ã‚‰åŸºæœ¬å½¢ã‚’å–å¾—
        if base in pn_dict:
            pn = float(pn_dict[base]) 
        else:
            pn = 'notfound'            # ãã®èªãŒPN Tableã«ãªã‹ã£ãŸå ´åˆ
        word['PN'] = pn
        diclist_new.append(word)
    return(diclist_new)

# å„ãƒ„ã‚¤ãƒ¼ãƒˆã®PNå¹³å‡å€¤ã‚’æ±‚ã‚ã‚‹
def get_mean(dictlist):
    pn_list = []
    for word in dictlist:
        pn = word['PN']
        if pn!='notfound':
            pn_list.append(pn)
    if len(pn_list)>0:
        pnmean = np.mean(pn_list)
    else:
        pnmean=0
    return pnmean

# å–å¾—ã—ãŸãƒ„ã‚¤ãƒ¼ãƒˆã®èª­ã¿è¾¼ã¿
df_tweets = pd.read_csv('./tweets.csv', names=['id', 'date', 'text', 'fav', 'RT'], index_col='date')
df_tweets = df_tweets.drop('text', axis=0)
df_tweets.index = pd.to_datetime(df_tweets.index)
df_tweets = df_tweets[['text']].sort_index(ascending=True)

# å²©æ³¢å›½èªè¾æ›¸ã®èª­ã¿è¾¼ã¿
pn_df = pd.read_csv('./6050_stock_price_prediction_data/pn_ja.csv', encoding='utf-8', names=('Word','Reading','POS', 'PN'))
#word_listã«ãƒªã‚¹ãƒˆå‹ã§Wordã‚’æ ¼ç´ã—ã¦ãã ã•ã„ã€‚
word_list = list(pn_df['Word'])
#pn_listã«ãƒªã‚¹ãƒˆå‹ã§PNã‚’æ ¼ç´ã—ã¦ãã ã•ã„ã€‚
pn_list   = list(pn_df['PN'])
#pn_dictã¨ã—ã¦word_list, pn_listã‚’æ ¼ç´ã—ãŸè¾æ›¸ã‚’ä½œæˆã—ã¦ãã ã•ã„ã€‚
pn_dict   = dict(zip(word_list, pn_list))


# means_listã¨ã„ã†ç©ºã®ãƒªã‚¹ãƒˆã‚’ä½œã‚Šãã“ã«ãƒ„ã‚¤ãƒ¼ãƒˆã”ã¨ã®å¹³å‡å€¤ã‚’æ±‚ã‚ã¦ãã ã•ã„ã€‚
means_list = []
for tweet in df_tweets['text']:
    dl_old = get_diclist(tweet)
    dl_new = add_pnvalue(dl_old, pn_dict)
    pnmean = get_mean(dl_new)
    means_list.append(pnmean)
df_tweets['pn'] = means_list
df_tweets =  df_tweets.resample('D').mean()

# æ—¥ä»˜ã‚’xè»¸PNå€¤ã‚’yè»¸ã«ã—ã¦ãƒ—ãƒ­ãƒƒãƒˆã—ã¦ãã ã•ã„ã€‚
x = df_tweets.index
y = df_tweets.pn
plt.plot(x,y)
plt.grid(True)

# df_tweets.csvã¨ã„ã†åå‰ã§df_tweetsã‚’å‡ºåŠ›ã—ã¦ãã ã•ã„ã€‚
df_tweets.to_csv('./df_tweets.csv')
```
